# Dist Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd



#dist plot is called distribution plot
#It is a combination of histogram and line plot or kernel density estimation(kde)
#histogram shows the distribution of data , while the line plot shows the density of the data.
#the density is a measure of how many data points lie in the given area
```

**Syntax**

```python
#dataset = sns.load_dataset(" ")

#sns.distplot(x = " ", y = " ", data = " dataset ")

# or sns.distplot(x = data["column_name1"] , y = data["column_name2"])

# or sns.distplot(x = " data.column_name" , y = "data.column_name" )
```

```python
var3 = sns.load_dataset("penguins")
var3
```

```python
#three ways to plot

sns.displot(x = "bill_length_mm",data = var3)
sns.displot(var3["bill_length_mm"])
sns.displot(var3.bill_length_mm)

plt.show()
```

```python
sns.displot(var3["flipper_length_mm"])
plt.grid(False)
plt.show()
```

```python
#Organising the x-tick markers using bins parameter
#dividing the data into ranges or intervals or bins , and 
#then counting the number of observations that fall into each range or internval or bin. 
#The height of each bar in a histogram corresponds to the number of data points that fall into the respective range or interval or bin.


sns.displot(var3["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240])
plt.grid(False)
plt.show()
```

```python
#kde is set false as default
#rug parameter denotes the scale line on the bottom axes


sns.displot(var3["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde = True , rug = True)
plt.grid(False)
plt.show()
```

```python
sns.displot(var3["flipper_length_mm"], bins = [170,180,190,200,210,220,230,240], kde = True , rug = True, color = "green")
plt.grid(False)
plt.show()
```

```python
sns.displot(var3["flipper_length_mm"], kde = True , rug = True, color = "green", log_scale = True)
plt.grid(False)
plt.show()
```
