```python
import folium
folium_map = folium.Map()
folium_map
```

```python
#Centering the map on Mumbai
#For that we need location of mumbai in term of latitude and longitude

location = [19.0760 , 72.8777 ]

folium_map = folium.Map(location = location,
                        zoom_start = 10,
                        zoom_control = False)
folium_map

```

```python
#setting initial zoom level to 10
#removing zoom control buttons on the map

location = [19.0760 , 72.8777 ]

folium_map = folium.Map(location = location,
                        zoom_start = 10,
                        zoom_control = False)
folium_map
```
